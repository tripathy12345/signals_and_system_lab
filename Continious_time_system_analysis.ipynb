Impulse responce, Step responce, Magnitude and Phase responce

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, step, impulse, bode

# Define the transfer function coefficients
numerator = [1]
denominator = [1, 0.5, 1]

# Create a TransferFunction object
system = TransferFunction(numerator, denominator)

# Impulse response analysis
t_impulse, y_impulse = impulse(system)
plt.figure(figsize=(8, 6))
plt.plot(t_impulse, y_impulse)
plt.title('Impulse Response')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Step response analysis
t_step, y_step = step(system)
plt.figure(figsize=(8, 6))
plt.plot(t_step, y_step)
plt.title('Step Response')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Frequency response analysis (Bode plot)
freq, mag, phase = bode(system)
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.semilogx(freq, mag)
plt.title('Bode Plot - Magnitude Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.semilogx(freq, phase)
plt.title('Bode Plot - Phase Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
