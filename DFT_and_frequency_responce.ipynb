*Discrete Fourier Transform of a signal*

```python
#import libraries
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Record the end time
end_time = time.time()
# Calculate the elapsed time
N=8
#discrete time sequence (DTS) (input from user)
x = np.array([1, 3, 2, 5, 6, 4])
print('Discrete Time sequence (x):',x) #Display the DTS x
N1 = len(x) #length of the DTS
x1 = np.append(x, np.zeros(N - N1))
wn = np.exp(-2j * np.pi / N)
X = np.zeros(N, dtype=complex)
print('DTS after zero padding (x1):',x1)
```

```python
#Calculate the DFT
start_time = time.time()
for k in range(N):
  for n in range(N):
    X[k] += x1[n] * (wn ** (k * n))
# Record the end time
end_time = time.time()
# Calculate the elapsed time
elapsed_time1 = end_time - start_time
print(f"For DFT elapsed time is:",elapsed_time1)
```

```python
start_time2 = time.time()
Y = np.fft.fft(x1)
# Record the end time
end_time2 = time.time()
# Calculate the elapsed time
elapsed_time2 = end_time2 - start_time2
print(f"For FFT elapsed time is:",elapsed_time2)
```

```python
#Plotting of DFT
X1 = np.real(X)
Y1 = np.imag(X)
plt.subplot(2, 1, 1)
plt.stem(X1, Y1)
plt.title('Discrete Fourier Transform using for loop')
plt.xlabel('Real Axis')
plt.ylabel('Imaginary Axis')
plt.subplots_adjust(hspace=1)
X2 = np.real(Y)
Y2 = np.imag(Y)
plt.subplot(2, 1, 2)
plt.stem(X2, Y2)
plt.title('Discrete Fourier Transform using FFT ')
plt.xlabel('Real Axis')
plt.ylabel('Imaginary Axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sinusoidal signal
fs = 250  # Sampling frequency in Hz
t = np.arange(0, 1, 1/fs)  # Time vector from 0 to 1 second with 1/fs increment
f_signal = 50  # Frequency of the sinusoidal signal in Hz
sinusoidal_signal = np.sin(2 * np.pi * f_signal * t)

# Plot the generated sinusoidal signal
plt.subplot(3, 1, 1)
plt.plot(t, sinusoidal_signal)
plt.title('Generated Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Calculate the FFT
n = len(sinusoidal_signal)
fft_result = np.fft.fft(sinusoidal_signal)
freq = np.fft.fftfreq(n, 1/fs)  # Frequency axis

# Plot the magnitude spectrum
plt.subplot(3, 1, 2)
plt.plot(freq, np.abs(fft_result))
plt.title('Magnitude Spectrum (FFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

# Plot the Phase spectrum
plt.subplot(3, 1, 3)
plt.plot(freq, np.angle(fft_result))
plt.title('Phase Angle')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase')

plt.tight_layout()
plt.show()
```
