###Butterworth lowpass filter using python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

f = 7
fs = 200
t = np.arange(0, 1, 1/fs)  # 0 to 1 sec, sampling time Ts = 1/Fs
signal = np.sin(2 * np.pi * f * t)

plt.subplot(4, 1, 1)
plt.plot(t, signal)
plt.title('x(t) signal')

noisySignal = signal + np.random.rand(len(signal))
plt.subplot(4, 1, 2)
plt.plot(t, noisySignal)
plt.title('Noisy signal')

# FFT
fftPoint = 2**np.ceil(np.log2(len(noisySignal)))
freDomain = np.abs(np.fft.fft(noisySignal, int(fftPoint)))  # signal in frequency domain
nFft = np.arange(fftPoint)
plt.subplot(4, 1, 3)
plt.plot(nFft, freDomain)
plt.title("Frequency")

# Filtering
order = 1
wc=8*2/fs
b, a = butter(order, wc, 'lowpass')
filteredSignal = lfilter(b, a, noisySignal)
plt.subplot(4, 1, 4)
plt.plot(t, filteredSignal)
plt.title("Signal after filtering")

plt.tight_layout()
plt.show()
```

##Butterworth bandpass filter using Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

f = 7
fs = 200
t = np.arange(0, 1, 1/fs)  # 0 to 1 sec, sampling time Ts = 1/Fs
signal = np.sin(2 * np.pi * f * t)

plt.subplot(4, 1, 1)
plt.plot(t, signal)
plt.title('x(t) signal')

noisySignal = signal + np.random.rand(len(signal))
plt.subplot(4, 1, 2)
plt.plot(t, noisySignal)
plt.title('Noisy signal')

# FFT
fftPoint = 2**np.ceil(np.log2(len(noisySignal)))
freDomain = np.abs(np.fft.fft(noisySignal, int(fftPoint)))  # signal in frequency domain
nFft = np.arange(fftPoint)
plt.subplot(4, 1, 3)
plt.plot(nFft, freDomain)
plt.title("Frequency")

# Filtering
order = 1
wl=4*2/fs
wh=11*2/fs
wc=[wl, wh]
b, a = butter(order, wc, 'bandpass')
filteredSignal = lfilter(b, a, noisySignal)
plt.subplot(4, 1, 4)
plt.plot(t, filteredSignal)
plt.title("Signal after filtering")

plt.tight_layout()
plt.show()
```

##Butterworth highpass filter using python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Generate a signal
fs = 1000  # Sampling frequency
t = np.linspace(0, 1, fs, endpoint=False)  # 1 second of data
f_signal = 10  # Frequency of the signal
signal = np.sin(2 * np.pi * f_signal * t)

# Add noise
np.random.seed(42)
noise = 0.5 * np.random.normal(size=len(t))
noisy_signal = signal + noise

# Design a Butterworth highpass filter
cutoff_frequency = 12  # Cutoff frequency in Hz
order = 4  # Filter order
b, a = butter(order, cutoff_frequency, btype='high', fs=fs)

# Apply the highpass filter
filter_output = lfilter(b, a, noisy_signal)
filtered_signal=noisy_signal-filter_output
# Plot the results
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, signal, label='Original Signal')
plt.title('Original Signal')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, noisy_signal, label='Noisy Signal')
plt.title('Noisy Signal')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.title('Filtered Signal')
plt.legend()

plt.tight_layout()
plt.show()
```

##Butterwoth bandstop filter

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Generate a signal
fs = 1000  # Sampling frequency
t = np.linspace(0, 1, fs, endpoint=False)  # 1 second of data
f_signal = 50  # Frequency of the signal
signal = np.sin(2 * np.pi * f_signal * t)

# Add noise
np.random.seed(42)
noise = 0.5 * np.random.normal(size=len(t))
noisy_signal = signal + noise

# Design a Butterworth bandstop filter
stop_band = [45, 55]  # Stop band frequencies in Hz
order = 4  # Filter order
b, a = butter(order, stop_band, btype='bandstop', fs=fs)

# Apply the bandstop filter
output_filter = lfilter(b, a, noisy_signal)
filtered_signal=noisy_signal-output_filter
# Plot the results
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, signal, label='Original Signal')
plt.title('Original Signal')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, noisy_signal, label='Noisy Signal')
plt.title('Noisy Signal')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(t, filtered_signal, label='Filtered Signal')
plt.title('Filtered Signal')
plt.legend()

```
