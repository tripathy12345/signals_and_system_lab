The objective of lab 1 is to generate different signals using python


1.1 Unit Sample sequence

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
y_axis = np.concatenate([np.zeros(N), [1], np.zeros(N)])
x_axis = np.arange(-N, N+1)

plt.stem(x_axis, y_axis, use_line_collection=True)
# If you want filled markers, you can use the following line instead
# plt.stem(x_axis, y_axis, markerfmt='rd', linefmt='-', basefmt=' ')

# Beautification
plt.xlabel('n')
plt.ylabel('x')
plt.axis([-12, 12, 0, 1.5])
plt.title('Unit Sample Sequence')

plt.show()
```

1.2 Unit step sequence

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10
x = np.concatenate([np.zeros(N), [1], np.ones(N)])
n = np.arange(-N, N+1)

plt.stem(n, x, use_line_collection=True)

# Beautification
plt.xlabel('n')
plt.ylabel('x')
plt.axis([-N-1, N+1, 0, 1.5])
plt.title('Unit Step Sequence')

plt.show()
```

1.3 Exponential Sequence

```python
import numpy as np
import matplotlib.pyplot as plt

N = 21
n = np.arange(0, N+1)  # Total Points (N+1)

xt_grow = 1.5**n  # Growing Sequence
xt_decay = 0.8**n  # Decaying Sequence

plt.subplot(2, 1, 1)
plt.plot(n, xt_grow)
plt.xlabel('n')
plt.ylabel('Growing x(n)')

plt.subplot(2, 1, 2)
plt.plot(n, xt_decay)
plt.xlabel('n')
plt.ylabel('Decaying x(n)')

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

1.5 Sinusoidal Signal

```python
import numpy as np
import matplotlib.pyplot as plt

# Formula: x(n) = A * sin(2 * pi * f * t + ph)

# Time vector
t = np.arange(0, 6.8, 0.1)  # End limit should contain at least one full cycle

# Amplitude
A = 5

# Frequency
f = 0.3  # Hz

# Phase
ph = 0

xt = A * np.sin(2 * np.pi * f * t + ph)

plt.plot(t, xt, '-', linewidth=3)

# Beautification
plt.grid(True)
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.legend(["Sinwave"])

plt.show()
```

1.6 Generate random signal

```python
import numpy as np
import matplotlib.pyplot as plt

# Random 10 point times
t = np.arange(0, 100.1, 0.1)

# Random Frequency
f = np.random.rand(1)

# Random Amplitude
A = np.random.rand(1)

xt = A * np.cos(2 * np.pi * f * t)

plt.plot(t, xt)
plt.grid(True)
plt.show()
```
